```python
def detecta_oulier(df,variable,registro):
    q3=df[variable].quantile(0.75)
    q1=df[variable].quantile(0.25)

    di=q3-q1

    outsupModerado1=q3+1.5*di
    outsupModerado2=q3+3*di

    outinfModerado1=q1-(1.5*di)
    outinfModerado2=q1-(3*di)

    filtrosup1=df[variable]>=outsupModerado1
    filtrosup2=df[variable]<=outsupModerado2
    
    filtroinf1=df[variable]<=outinfModerado1
    filtroinf2=df[variable]>=outinfModerado2

    filtrosup3=df[variable]>outsupModerado2
    filtroinf3=df[variable]<outinfModerado2

    if len(df[filtrosup1 & filtrosup2])>0:
        print("outliers moderados q3:",len(df[filtrosup1 & filtrosup2]))
        display(df[filtrosup1 & filtrosup2][[variable,registro]])
    if len(df[filtrosup3])>0:
        print("outliers severos q3:",len(df[filtrosup3]))
        display(df[filtrosup3][[variable,registro]])
    if len(df[filtroinf1 & filtroinf2])>0:
        print("outliers moderados q1:", len(df[filtroinf1 & filtroinf2]))
        display(df[filtroinf1 & filtroinf2][[variable,registro]])
    if len(df[filtroinf3]):
        print("outliers severos q1:",len(df[filtroinf3]))
        display(df[filtroinf3][[variable,registro]])
    return(df[filtrosup1 & filtrosup2],df[filtrosup3],
           df[filtroinf1 & filtroinf2],df[filtroinf3]
          )
```

```python
print("Razonamiento cuantitativo")    
sevq3,modq3,modq1,sevq3=detecta_oulier(datos,'MOD_RAZONA_CUANTITAT_PUNT','ESTU_CONSECUTIVO')
```
